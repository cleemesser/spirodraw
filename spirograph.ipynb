# Understanding how a spirograph works using python
The spirograph consists of geared discs and geared rings. This allows you to draw a cool looking set of curves in the plane.
These are called [hypocycloids](https://en.wikipedia.org/wiki/Hypocycloid).

These hypocycloids are defined by the radius of the smaller circle which rotates inside the larger circle defined by the geared ring.

These are parameterized functions:
    $$ x(t) = (a-b) \cos(t) + b \cos((a-b)t/b) $$
    $$ y(t) = (a-b) \sin(t) - b \sin((a-b)t/b) $$

They are implemented in the spirodraw package in spirodraw.spirograph_functions.py

To get started using the package in the jupyter notebook, first import some of the drawing functions. Here we will use matplotlib as a simple drawing method:

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline  
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (7,7)
plt.rcParams["figure.dpi"] = 300

from spirodraw.drawing_mpl import (
    start_drawing,
    add_hypocycloid,
    interactive_spiro,
    example_spiro1,
    example_spiro2,
    example_random_spiro3,
) 
```

Here is an example of something you can draw:

```python
example_spiro1()
```

Note that this uses (5,2) and (10, 2.2) as the radii of the circles.
Here is how we can reproduce this using simplier functions:

```python
fig, ax = start_drawing()
ax = add_hypocycloid(ax, big_radius=5, little_radius=2)
```

```python
add_hypocycloid(ax, big_radius=10, little_radius=2.2) # note you do not need to restrict yourself to integers
add_hypocycloid(ax, big_radius=15, little_radius=2.13)
fig # you can show the figure again by making it the last thing in a cell
```
